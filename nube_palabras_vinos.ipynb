```python
##Create stopword list:
stopwords = set(STOPWORDS)
stopwords.update(["drink", "now", "wine", "flavor", "flavors"])

# Generate a word cloud image
wordcloud = WordCloud(stopwords=stopwords, background_color="white").generate(text)

# Display the generated image:
# the matplotlib way:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
wine_mask = np.array(Image.open("images/wine_mask_xgk1tq.png"))
wine_mask
```

```python
def transform_format(val):
    if val == 0:
        return 255
    else:
        return val
```

```python
# Transform your mask into a new one that will work with the function:
transformed_wine_mask = np.ndarray((wine_mask.shape[0],wine_mask.shape[1]), np.int32)

for i in range(len(wine_mask)):
    transformed_wine_mask[i] = list(map(transform_format, wine_mask[i]))
```

```python
# Check the expected result of your mask
transformed_wine_mask
```

```python
# Create a word cloud image
wc = WordCloud(background_color="white", max_words=1000, mask=transformed_wine_mask,
               stopwords=stopwords, contour_width=3, contour_color='firebrick')
```

```python
# Generate a wordcloud
wc.generate(text)

# store to file
wc.to_file("images/wine.png")

# show
plt.figure(figsize=[20,10])
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
# Join all reviews of each country:
usa = " ".join(review for review in df[df["country"]=="US"].description)
fra = " ".join(review for review in df[df["country"]=="France"].description)
ita = " ".join(review for review in df[df["country"]=="Italy"].description)
spa = " ".join(review for review in df[df["country"]=="Spain"].description)
por = " ".join(review for review in df[df["country"]=="Portugal"].description)
```

```python
def generate_world_cloud_by_country(country, custom_background_color):
    custom_background_color = custom_background_color
    if country=="portugal":
        mask = np.array(Image.open("images/"+country+".png"))
        por = " ".join(review for review in df[df["country"]=="Portugal"].description)
        wordcloud = WordCloud(stopwords=stopwords, background_color=custom_background_color, max_words=1000, mask=mask).generate(por)
    if country=="france":
        mask = np.array(Image.open("images/"+country+".png"))
        fra = " ".join(review for review in df[df["country"]=="France"].description)
        wordcloud = WordCloud(stopwords=stopwords, background_color=custom_background_color, max_words=1000, mask=mask).generate(fra)
    if country=="us":
        mask = np.array(Image.open("images/"+country+".png"))
        usa = " ".join(review for review in df[df["country"]=="US"].description)
        wordcloud = WordCloud(stopwords=stopwords, background_color=custom_background_color, max_words=1000, mask=mask).generate(usa)
    if country=="italy":
        mask = np.array(Image.open("images/"+country+".png"))
        ita = " ".join(review for review in df[df["country"]=="Italy"].description)
        wordcloud = WordCloud(stopwords=stopwords, background_color=custom_background_color, max_words=1000, mask=mask).generate(ita)
    if country=="spain":
        mask = np.array(Image.open("images/"+country+".png"))
        spa = " ".join(review for review in df[df["country"]=="Spain"].description)
        wordcloud = WordCloud(stopwords=stopwords, background_color=custom_background_color, max_words=1000, mask=mask).generate(spa)

    # create coloring from image
    image_colors = ImageColorGenerator(mask)
    plt.figure(figsize=[7,7])
    plt.imshow(wordcloud.recolor(color_func=image_colors), interpolation="bilinear")
    plt.axis("off")
```

```python
generate_world_cloud_by_country("portugal", "white")
generate_world_cloud_by_country("france", "white")
generate_world_cloud_by_country("us", "white")
generate_world_cloud_by_country("italy", "white")
generate_world_cloud_by_country("spain", "white")
```
